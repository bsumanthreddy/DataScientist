## EDA on Water Pump Dataset

```python
# import necessary libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os

# visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data

sensor_data = pd.read_csv('/content/sensor.csv')
```

```python
sensor_data.head()
```

```python
# see the labels we have
sensor_data['machine_status'].unique()
```

```python
sensor_data.describe()
```

```python
sensor_data.describe()
```

```python
# get further info from dataset
sensor_data.info()
```

```python
# plot a heatmap of the missing values
plt.figure(figsize=(20,5))
sns.heatmap(sensor_data.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.title('Missing Data Heatmap')
plt.show();
```

```python
# pie chart of machine status
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,5))
sensor_data.groupby('machine_status').size().plot(kind='bar', ax=ax1, color=['black', 'cornflowerblue', 'gold'])
sensor_data.groupby('machine_status').size().plot(kind='pie', ax=ax2, autopct='%.4f%%', colors=['black', 'cornflowerblue', 'gold'])
ax1.set_title('Machine Status Bar')
ax1.set_xlabel('')
ax2.set_title('Machine Status Pie')
ax2.set_ylabel('')
plt.show();
```

```python
ax = sns.countplot(x="timestamp", hue="machine_status", data=sensor_data)
plt.show()
```
